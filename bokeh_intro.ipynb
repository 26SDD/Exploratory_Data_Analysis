```python
from bokeh.charts import Scatter, output_notebook, show
import pandas as pd
import numpy as np
from scipy import stats
from pandas import Series, DataFrame
from bokeh.io import output_notebook
output_notebook()
```

```python
#polynomial data generation
x = np.random.uniform(10000,20000, size=(64,))
x = x.astype(int)
print "x values are"
print x
print
y = .0005 * x**2 +.01*x + 20
#y = 2*x + 20
print "equation fitted is .5 * x**2 +4*x + 20"
print "y values are"
print y

x1 = range(min(x),max(x),int((max(x)-min(x))/200))
y_p = [int(.0005 * i**2 +.01*i) + 20 for i in x1]
y_l = [int((14.946 * i) -107616.96288 ) for i in x1]
y_po = [int(0.00050897586 * (i ** 1.9983) ) for i in x1]
```

```python
from pandas.tseries.offsets import *
dates_index =pd.date_range('2016-01-04', '2016-03-31', freq=BDay())
dates_index
```

```python
from bokeh.plotting import figure, output_file,output_notebook, show
#from bokeh.models import Axis

output_notebook()

p=figure(plot_width=800,plot_height=800,logo=None,responsive= False)
df = pd.DataFrame({'sale_volume':x, 'short_volume':y, 'tdate':dates_index})
df.head(3)

#p.title="Temperature and Air Pressure"
p.title.text_color="Gray"
p.title_text_font="arial"
p.title_text_font_style="bold"
p.xaxis.minor_tick_line_color=None
p.yaxis.minor_tick_line_color=None
p.xaxis.axis_label="sale_volume"
p.yaxis.axis_label="short_volume"   
p.left[0].formatter.use_scientific = False

p.circle(df["sale_volume"],df["short_volume"],size=10)

# add polynomial
p.line(x1, y_p, legend="poly", line_width=4,line_color="red")
p.line(x1, y_l, legend="linear", line_width=2, line_color="purple")
p.line(x1, y_po, legend="power", line_width=3,line_color="orange",line_dash="4 4")

output_file("plots.html")

show(p)
```

```python
from scipy.optimize import curve_fit

x = np.array(df.sale_volume)
y = np.array(df.short_volume)
#l_x = np.array(np.log(df_31194.sale_volume))
#l_y = np.array(np.log(df_31194.sale_volume))


def fit_func_poly(x,a,b,c):
    return a*(x**2) + b*x + c #polynomial

def fit_func_exp(x, a, b):
    return a * np.e**(b * x)

def fit_func_linear(x, a, b):
    return a * x + b

def fit_func_log (x, a, b):
    return np.log(a) + b*np.log(x)

def fit_func_power (x, a, b):
    return a* x**b

p_linear = curve_fit(fit_func_linear, x, y,maxfev=20000)
print "linear",['{:.20f}'.format(t) for t in p_linear[0]]
p_poly = curve_fit(fit_func_poly, x, y,maxfev=20000)
print "poly",['{:.20f}'.format(t) for t in p_poly[0]]
p_power = curve_fit(fit_func_power, x, y,maxfev=20000)
print "power",['{:.20f}'.format(t) for t in p_power[0]]
p_log = curve_fit(fit_func_log, x, y,maxfev=20000)
print "log",['{:.30f}'.format(t) for t in p_log[0]]
#p_exp = curve_fit(fit_func_exp, x, y,maxfev=20000,x=(1, 1e-6, 1))
#print p_exp
```
